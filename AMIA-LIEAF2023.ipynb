```python

from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score, roc_curve, auc

y_test_one_hot = label_binarize(y_test, classes=rfc.classes_)

fpr = {}
tpr = {}
roc_auc = {}

# Get the probabilities for each class
rfc_y_proba = rfc.predict_proba(X_test)

# Calculate AUROC for each class
for i in range(len(rfc.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_test_one_hot[:, i], rfc_y_proba[:, i])  # Use probabilities for the current class
    roc_auc[i] = auc(fpr[i], tpr[i])

# Print AUROC for each class
for i in range(len(rfc.classes_)):
    print(f"Class {rfc.classes_[i]}: AUROC = {roc_auc[i]}")

classes = rfc.classes_
print("Classes:", classes)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Collect predictions and true labels for each class
y_true = np.array([0, 1, 0, 2, 1, 2])  # True labels (ground truth)
y_scores = np.array([
    [0.9, 0.1, 0.2],  # Predicted probabilities for class 0
    [0.4, 0.6, 0.3],  # Predicted probabilities for class 1
    [0.2, 0.8, 0.5],  # Predicted probabilities for class 0
    [0.3, 0.4, 0.9],  # Predicted probabilities for class 2
    [0.7, 0.3, 0.2],  # Predicted probabilities for class 1
    [0.1, 0.2, 0.8],  # Predicted probabilities for class 2
])

# Compute ROC curve for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = y_scores.shape[1]

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true == i, y_scores[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Calculate macro-average ROC curve
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)

for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= n_classes

# Calculate macro-average AUC
mean_auc = auc(all_fpr, mean_tpr)

# Plot the ROC curve
plt.figure()
plt.plot(all_fpr, mean_tpr, label=f'Macro-average ROC (AUC = {mean_auc:.2f})', color='b', linestyle='--')
plt.plot([0, 1], [0, 1], color='r', linestyle=':', label='Random Guessing')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Macro-average ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
#FEATURE IMPORTANCE
importances = rfc.feature_importances_
# Sort the feature importance in descending order
sorted_indices = np.argsort(importances)[::-1]
import matplotlib.pyplot as plt
 
plt.title('RF Feature Importance')
plt.bar(range(X_train.shape[1]), importances[sorted_indices], align='center')
plt.xticks(range(X_train.shape[1]), X_train.columns[sorted_indices], rotation=90)
plt.tight_layout()
plt.show() 
```
